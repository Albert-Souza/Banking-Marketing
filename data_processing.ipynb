```python
import pandas as pd
```

```python
df = pd.read_csv('Data/bank-additional-full.csv', sep=';')
```

```python
df.info()
```

```python
df2 = df.drop(columns=df.columns[15:20])
df2.head()
```

```python
df3 = df2.rename({
    'age': 'Idade',
    'job': 'OcupaÃ§Ã£o',
    'marital': 'Estado Civil',
    'education': 'Escolaridade',
    'default': 'Default',
    'housing': 'EmprÃ©stimo ImobiliÃ¡rio',
    'loan': 'EmprÃ©stimo Pessoal',
    'contact': 'Contato',
    'month': 'MÃªs',
    'day_of_week': 'Dia da Semana',
    'duration': 'DuraÃ§Ã£o',
    'campaign': 'Campanha',
    'pdays': 'Dias Passados',
    'previous': 'Contatos Anteriores',
    'poutcome': 'Resultado Anterior',
    'y': 'Resultado'
}, axis=1)

df3.head()
```

```python
df4 = df3.copy()

df4['OcupaÃ§Ã£o'] = df4['OcupaÃ§Ã£o'].replace({
    "admin.": "Administrativo",
    "unknown": "Desconhecido",
    "unemployed": "Desempregado",
    "management": "GerÃªncia",
    "housemaid": "Empregada DomÃ©stica",
    "entrepreneur": "Empreendedor",
    "student": "Estudante",
    "blue-collar": "OperÃ¡rio",
    "self-employed": "AutÃ´nomo",
    "retired": "Aposentado",
    "technician": "TÃ©cnico",
    "services": "ServiÃ§os"
})

df4['Estado Civil'] = df4['Estado Civil'].replace({
    'married': 'Casado',
    'single': 'Solteiro',
    'divorced': 'Divorciado',
    "unknown": "Desconhecido"
})

df4['Escolaridade'] = df4['Escolaridade'].replace({
    "basic.4y": "Fundamental (4Âº ano)",
    "basic.6y": "Fundamental (6Âº ano)",
    "basic.9y": "Fundamental (9Âº ano)",
    "high.school": "Ensino MÃ©dio",
    "illiterate": "Analfabeto",
    "professional.course": "Curso Profissionalizante",
    "university.degree": "Ensino Superior",
    "unknown": "Desconhecido"
})

df4['Default'] = df4['Default'].replace({
    'yes': 'Sim',
    'no': 'NÃ£o',
    'unknown': 'Desconhecido'
})

df4['EmprÃ©stimo ImobiliÃ¡rio'] = df4['EmprÃ©stimo ImobiliÃ¡rio'].replace({
    'yes': 'Sim',
    'no': 'NÃ£o',
    'unknown': 'Desconhecido'
})

df4['EmprÃ©stimo Pessoal'] = df4['EmprÃ©stimo Pessoal'].replace({
    'yes': 'Sim',
    'no': 'NÃ£o',
    'unknown': 'Desconhecido'
})

df4['Contato'] = df4['Contato'].replace({
    "telephone": "Telefone Fixo",
    "cellular": "Celular",
    "unknown": "Desconhecido"
})

df4['Resultado Anterior'] = df4['Resultado Anterior'].replace({
    'success': 'Sucesso',
    'failure': 'Falha',
    'nonexistent': 'NÃ£o Existente'
})

df4['Resultado'] = df4['Resultado'].replace({
    'yes': 'Sucesso',
    'no': 'Falha'
})

df4['MÃªs'] = df4['MÃªs'].replace({
    "jan": '1',
    "feb": '2',
    "mar": '3',
    "apr": '4',
    "may": '5',
    "jun": '6',
    "jul": '7',
    "aug": '8',
    "sep": '9',
    "oct": '10',
    "nov": '11',
    "dec": '12'
}).astype(int)

df4['Dia da Semana'] = df4['Dia da Semana'].replace({
    "sun": '1',
    "mon": '2',
    "tue": '3',
    "wed": '4',
    "thu": '5',
    "fri": '6',
    "sat": '7'
}).astype(int)

df4.head()
```

```python
df5 = df4.copy()
df5['Dias Passados'] = df5['Dias Passados'].replace({999: -1})

df5.head()
```

```python
df5.to_excel('Data/processed_data.xlsx', index=False)
```
